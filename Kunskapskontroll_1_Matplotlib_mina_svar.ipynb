### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4)
y = x**2
```

```python
plt.plot(x,y)
plt.title('y = x^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
fig, y_x = plt.subplots()
y_x.plot(x,y)
y_x.set_xlabel('X-axis')
y_x.set_ylabel('Y-axis')
y_x.set_title('y = x^2')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (scatt, barp) = plt.subplots(2,1, figsize=(9, 14)) #Creating the figure with two subplots scatt and barp
#Also makes the dimensions 2,1 which means they are stacked horizontally, with the size 8 inch tall and 10 inch wide.

scatt.scatter(x, y, color='r', label='np.linspace')
scatt.scatter(random_data_x, random_data_y, color='b', label='np.random.seed')
scatt.set_title('Scatter Plot')
scatt.set_xlabel('X-axis')
scatt.set_ylabel('Y-axis')
scatt.legend() #To make it clear to the viewer what the different colors represent.

barp.bar(names, values)
barp.set_xlabel('Fruit names') 
barp.set_ylabel('Fruit quantity')
barp.set_title('Fruits')

plt.tight_layout() #making it tidy and nice
plt.show() #showing it
```
